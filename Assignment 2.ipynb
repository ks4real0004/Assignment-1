```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

#load dataset
digit = load_digits()

#dataset information
print("Dataset Shape:",digit.data.shape)
print("Image Shape:",digit.images.shape)
print("Image Labels:",digit.target_names)

#display the first image
plt.figure(figsize=(4,4))
plt.imshow(digit.images[5], cmap='pink')
plt.title(f"Label: {(digit.target[5])}")
plt.axis('off')
plt.show()
```

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

#load dataset
digits = load_digits()

fig,axes=plt.subplots(2,5,figsize = (10,5))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='coolwarm')
    ax.set_title(f"Digit: {digits.target[i]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
